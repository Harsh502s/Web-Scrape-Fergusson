```python
url = r"https://www.airlinequality.com/airline-reviews/air-india/"
response = get(url)
soup = BeautifulSoup(response.text, 'lxml')
print(soup.prettify())
```

```python
article_soup = soup.find_all('article', attrs={'class':"comp comp_reviews-airline querylist position-content"})
article_soup
```

```python
all_title = [i.find_all('div', attrs={'class':'body'}) for i in article_soup ]
titles = [i.find('h2').text for i in all_title[0]]
titles
# all_body = [i.find('div' , attrs={'class':'body','id':'anchor888660'}).find('div' , class_='tc_mobile') for i in article_soup ]
```

```python
all = [i.find_all('div', attrs={'class':'body'}) for i in article_soup ]
all_body = [i.find('div' , class_='tc_mobile').text for i in all[0] ]
print(len(all))
all_body
```

```python
data_dict = {'Title':titles, 'Body':all_body}
df = pd.DataFrame(data_dict)
df
```
